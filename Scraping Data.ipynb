```python
name_list = []
link_list = []
    
# Get the links and names for each user page
links = driver.find_elements_by_xpath("//div[@class='name-location']//a[@href]")

# Add names and links to their respective lists
for link in links:
    name_list.append(link.text)
    link_list.append(link.get_attribute('href'))
        
print(name_list)
print(link_list)
```

```python
location_list = []
climb_type_list = []
gender_list = []
age_list = []
        
# Get the location, gender, age, and type of climbing of each user
locations_genders_ages = driver.find_elements_by_xpath("//td[@class='small text-nowrap']")

for info in locations_genders_ages:
    info = info.text
    info = info.split('\n')
    location = info[0]
    climb_type = info[2]
    gender_age = info[1].split(' ')
    gender = gender_age[0]
    gender = gender.split(',')
    gender = gender[0]
    age = gender_age[1]
    location_list.append(location)
    climb_type_list.append(climb_type)
    gender_list.append(gender)
    age_list.append(age)
            
print(location_list)
print(climb_type_list)
print(gender_list)
print(age_list)
```

```python
def scrape(min_age, max_age, miles, zip_code, filename):
    
    df = pd.DataFrame(columns = ['Name', 'Link', 'Date', 'Gender', 'Age', 'Lead Follow',
                             'Membership', 'Climb Type', 'Best Times', 'Other Interests',
                             'Other Info', 'Route', 'Rating', 'Route Type',
                             'Style', 'Lead Style', 'Length', 'Notes', 
                             'URL', 'Pitches', 'Location','Avg Stars', 'Your Stars',
                            'Your Rating', 'Rating Code'])
    
    # Scrape each age
    for age in range(41, 42):

        # Create a driver
        driver = webdriver.Chrome(executable_path = '/Users/kaitlynzeichick/Downloads/chromedriver')
        driver.get("https://www.mountainproject.com/partner-finder")
        
        driver.implicitly_wait(0.5)
        
        age_plus_one = str(age+1)
        age = str(age)
    
        # Select the minimum age
        sel = Select(driver.find_elements_by_xpath("//select[@class='form-control']")[0])
        sel.select_by_visible_text(min_age) # change back to 'age'
        
        # Select the maximum age
        sel = Select(driver.find_elements_by_xpath("//select[@class='form-control']")[1])
        sel.select_by_visible_text(max_age) # change back to 'age_plus_one'
        
        # Select 100 miles
        sel = Select(driver.find_elements_by_xpath("//select[@class='form-control']")[2])
        sel.select_by_visible_text(miles)
        
        # Pick a zip code
        inputElement = driver.find_element_by_xpath("//input[@class='form-control inline-block']")
        inputElement.send_keys(zip_code)
        
        
        # Click Find Partners
        driver.find_element_by_xpath("//button[@class='btn btn-primary']").click()
        
        # GET NAMES AND LINKS
        name_list = []
        link_list = []
    
        # Get the links and names for each user page
        links = driver.find_elements_by_xpath("//div[@class='name-location']//a[@href]")

        # Add names and links to their respective lists
        for link in links:
            name_list.append(link.text)
            link_list.append(link.get_attribute('href'))
        
        #print(name_list)
        #print(link_list)
            
        # GET LOCATION, CLIMB TYPE, GENDER, AND AGE
        location_list = []
        climb_type_list = []
        gender_list = []
        age_list = []
        
        # Get the location, gender, age, and type of climbing of each user
        locations_genders_ages = driver.find_elements_by_xpath("//td[@class='small text-nowrap']")

        for info in locations_genders_ages:
            try:
                info = info.text
                info = info.split('\n')
                location = info[0]
                climb_type = info[2]
                gender_age = info[1].split(' ')
                gender = gender_age[0]
                gender = gender.split(',')
                gender = gender[0]
                age = gender_age[1]
                location_list.append(location)
                climb_type_list.append(climb_type)
                gender_list.append(gender)
                age_list.append(age)
            except:
                gender = 'Unknown'
                location = 'Unknown'
                age = 0
                climb_type = 'Unknown'
                gender_list.append(gender)
                location_list.append(location)
                age_list.append(age)
                climb_type_list.append(climb_type)
            
        # GET LEAD FOLLOW INFORMATION      
        lead_follow_list = []
        # Get lead and following capabilities for each user
        leads_follows = driver.find_elements_by_xpath("//td[@class='small text-nowrap hidden-sm-down']")

        for lead_follow in leads_follows:
            lead_follow_list.append(lead_follow.text)
            
        # GET EXTRA INFO, BEST TIMES, AND OTHER INTERESTS
        extra_info_list = []
        bestTimes_otherInterests = driver.find_elements_by_xpath("//td[@class='small hidden-md-down']")

        for bestTime_otherInterest in bestTimes_otherInterests:
            extra_info_list.append(bestTime_otherInterest.text)
        
        bestTimes_list = extra_info_list[::3]
        otherInterests_list = extra_info_list[1::3]
        otherInfo_list = extra_info_list[2::3]
        
        # GET MEMBERSHIP AND USER INFO
        count = 0
        for link in link_list:
            print('navigating to: ' + link)
            driver.get(link)
            
            membership = driver.find_element_by_xpath("//div[@class='info mt-1']")
            membership = membership.text
            membership = membership.split("Since")[1]
            membership = membership.split("Last")[0]
            
            # Click View All Ticks
            link = driver.find_elements_by_xpath("//span[@class='font-body pl-1']//a[@href]")

            # Try to expand Ticks, usually at link[1]
            try:
                link[1].click()

            # There might not be a to-do list, in which case the ticks are at link[0]    
            except:
            # There might not be a tick list. If there isn't, move on
                try:
                    link[0].click()
                except:
                    pass
                
            # Make sure that the title is Ticks.
            try:
                title = driver.find_element_by_xpath("//h1[@class='inline-block mt-1']")
                title = title.text
                title = title.split(">")[1]
                title = title.split("List")[0]

                if title != ' To-Do ':
                    # Get the link from the 'Export csv' button
                    link = driver.find_element_by_xpath("//span[@class='pl-2 font-body']//a[@href]")

                    csv_url = link.get_attribute('href')
                
                    # Get a requests.Response object
                    req = requests.get(csv_url)

                    # Generate the content from the requests.Response object
                    url_content = req.content

                    # Open a file object called downloaded.csv to write in
                    csv_file = open('downloaded.csv', 'wb')

                    # Write the the contents of the csv file stored at the URL to the downloaded.csv file
                    csv_file.write(url_content)

                    # Stop writing
                    csv_file.close()
                
                    # Read in the csv file        
                    df_csv = pd.read_csv('downloaded.csv')
                
                    df_csv['Name'] = name_list[count]
                    df_csv['Link'] = link_list[count]
                    df_csv['Membership'] = membership
                    df_csv['Best Times'] = bestTimes_list[count]
                    df_csv['Other Interests'] = otherInterests_list[count]
                    df_csv['Other Info'] = otherInfo_list[count]
                    df_csv['Location'] = location_list[count]
                    df_csv['Climb Type'] = climb_type_list[count]
                    df_csv['Gender'] = gender_list[count]
                    df_csv['Age'] = age_list[count]
                    df_csv['Lead Follow'] = lead_follow_list[count]
                    
                    df = df.append(df_csv, ignore_index=True)
                
            except:
                pass
            
            driver.back()
            
            count += 1  
        
        driver.close()
        
    df.to_csv(filename)
        
    return df
        
```
